```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('blinkit_customers.csv')
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
print(df.dtypes)

```

```python
df['registration_date'] = pd.to_datetime(df['registration_date'])
```

```python
sns.countplot(x='customer_segment', data=df)
plt.title('Distribution of Customer Segments')
plt.xlabel('Customer Segment')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_orders'], bins=30, kde=True)
plt.title('Total Orders Distribution')
plt.xlabel('Total Orders')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='total_orders', y='avg_order_value', data=df)
plt.title('Total Orders vs Average Order Value')
plt.xlabel('Total Orders')
plt.ylabel('Average Order Value')
plt.show()
```

## Insights
- Most customers belong to the Regular segment.
- The average number of orders is around 10 , with a few high-value outliers.
